## **Step 1: Deciding on a model and the setup**
I chose YOLOv8n.pt (pre-trained) model because it has YOLO's object detection algorith is renowned for fast image processing results and improved accuracy. YOLO v8 was launched in January and has a high rate of accuracy vs. previous models and has a well-structured Python package.

```python
#Install YOLOv8 from pip. Ultralytics YOLO comes with a pythonic Model and Trainer interface
!pip install ultralytics &> /dev/null

#To use YOLOv8 on Python and display the result, you will need the following libraries:
from ultralytics import YOLO
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import cv2

#And if you are on Google Colab also import this one:
from google.colab.patches import cv2_imshow
```

## **Step 2: Loading and using the model**

**Load the model:**

```python
#Load a pre-trained version of YOLOv8 (by default ultralytics gives us the most recent one):
model = YOLO("yolov8n.pt")
```

**Results after feeding an image into the model:**

```python
#Loading an image from the internet as URL or from directory path:
#https://images.unsplash.com/photo-1600880292203-757bb62b4baf?ixlib=rb-4.0.3&ixid=MnwxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8&auto=format&fit=crop&w=2070&q=80

image_path = str(input("Image's URL or directory path:")).strip() #strip remove spaces at the beginning and at the end of the string

#Transforming the image into a numpy array:
response = requests.get(image_path)
image = Image.open(BytesIO(response.content))
image = np.asarray(image)

#Run the prediction on our image (this should be fast, even without GPU):
if 'https://' in image_path or 'http://' in image_path:
  results = model.predict(image) #to run prediction on image with http prefix
else:
  results = model.predict(source='{}'.format(image_path)) #to run prediction on image from file directory
```

Comment:
We get a lot of objects. Some are relevant, and some are irrelevant - like the chairs.

The YOLOv8n.pt (pre-trained) model is able to detect all objects in the image. But it wasn't able to detect all images with the same confidence score. We can filter these objects according to this score.

For example, we can display only the objects with a confidence score higher than 85%. Thus, all the objects detected with a lower score will not be displayed. To filter according to the confidence score, we indicate conf=0.85.

**Results after filtering these objects to confidence score of 85%:**

```python
response = requests.get(image_path)
image = Image.open(BytesIO(response.content))
image = np.asarray(image)

#Run the prediction on our image where conf=0.85:
if 'https://' in image_path or 'http://' in image_path:
  results = model.predict(image, conf = 0.85) #to run prediction on image with http prefix
else:
  results = model.predict(source='{}'.format(image_path), conf = 0.85) #to run prediction on image from file directory
```

With a confidence threshold of 0.85, YOLOv8n was able to detect 2 persons.
